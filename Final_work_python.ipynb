> В ячейке ниже представлен код генерирующий **DataFrame**, которая состоит всего из 1 столбца. Наша задача перевести его в **one hot** вид и сделать без **get_dummies**.

#### Подгружаем необходимые библиотеки.

````python
# Библиотека для работы с табличными данными
import pandas as pd
# Библиотека для вычислений линейной алгебры
import numpy as np
# Библиотеки для рандомных значений
import random
````

#### Создаем колонку **'whoAmI'**, в которой будут рандомно генерироваться значения **'robot'** и **'human'**.

````python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

####Переводим **'DataFrame'** в **'One hot'** без функции **'get_dummies'**. 


> Создаем дополнительно две колонки **'human'** и **'robot'**. Где будут совпадения по названиям выводится **'1'**, иначе **'0'**.



````python
data['robot'] = np.where(data['whoAmI'] == 'robot', '1', '0')
data['human'] = np.where(data['whoAmI'] == 'human', '1', '0')
data
````
